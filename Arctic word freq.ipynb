```python
import xml.etree.ElementTree as ET
import string
import operator
import pandas as pd
```

Main function that count words:

```python
def word_count(text_data, wordFreq, punc):
    '''
    text_data - list of strings
    wordFreq  - list with frequent words to exlude
    punc      - string with punctuation to exclude
    '''
    words={}

    for line in text_data:
        splited = line.split()
        for ww in splited:
            word = ww.lower()
            word = ''.join((x for x in word if x not in punc))
        
            popWords= word in wordFreq
            if popWords == False:
                if words.has_key(word):
                    words[word]=words[word]+1
                else:
                    words[word]=1
    return words
```

Here I define some punctuation that I would like to exclude:

```python
punc=set(",./;'?&\'\"")
```

We also don't want to get a lot of very popular English words in the top, so I download the list of the [top 100 popular words](http://www.englishclub.com/vocabulary/common-words-100.htm) and add some that are specific for scientific vocabulary.

```python
fw = open('frequent_words_less.txt')
wordFreq=fw.readlines()
for ii, aa in enumerate(wordFreq):
    wordFreq[ii]=wordFreq[ii].rstrip()
```

Here I get abstracts from xml files, count words and put them in to the pandas data frame. 

```python
first = True
for year in range(2003,2013):
    tree = ET.parse('./xml/Arctic'+str(year)+'.xml')
    root = tree.getroot()
    text_data = []
    for paper in root:
        abstract = paper.findall('{http://ads.harvard.edu/schema/abs/1.1/abstracts}abstract')
        text_data.append(abstract[0].text.replace('\n', ' '))
    wordYear = word_count(text_data, wordFreq, punc)
    if first:
        data = pd.DataFrame({str(year):wordYear})
        
        print('First year '+str(year)+' is processed')
    else:
        data_temp = pd.DataFrame({str(year):wordYear})
        
        data = pd.concat([data, data_temp],axis=1)
        
        print('Year '+str(year)+' is processed')
        del data_temp
    first = False
        
    
        
```

I was to lazy to implement automatic division by number of papers, so made it here manually.

```python
data['2003'] = data['2003']/1173.
data['2004'] = data['2004']/1088.
data['2005'] = data['2005']/1029.
data['2006'] = data['2006']/1151.
data['2007'] = data['2007']/1186.
data['2008'] = data['2008']/1384.
data['2009'] = data['2009']/1953.
data['2010'] = data['2010']/1982.
data['2011'] = data['2011']/1656.
data['2012'] = data['2012']/1466.
```

Sort the data:

```python
dataSorted = pd.DataFrame(index=range(500))
for year in range(2003,2013):
    aa = data[str(year)].copy()
    aa = aa.order(ascending=False)
    dataSorted[str(year)] = aa[0:500].index
```

## Here is top 50 words by year:

```python
print(dataSorted.head(50))
```

Let's have a look at how word positions change between the first and the last years of our analysis.

Calculate differences:

```python
wordDiff = {}
for i in range(50):
    wword = dataSorted['2012'][i]
    
    difference = int(dataSorted['2003'][dataSorted['2003']==wword].index) - int(dataSorted['2012'][dataSorted['2012']==wword].index)
    wordDiff[wword]=difference
```

Create pandas data frame for convenience:

```python
dataWordDiff = pd.DataFrame({'Difference':wordDiff})
dataWordDiff['Difference'] = dataWordDiff['Difference'].astype('int')
dataWordDiffAbs = abs(dataWordDiff['Difference'])
```
